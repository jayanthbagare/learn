```python
import csv
import datetime as dt
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
```

```python
dates = []
prices = []
```

```python
def get_data(filename):
    with open(filename,'r') as csvfile:
        csvFileReader = csv.reader(csvfile)
        next(csvFileReader)
        
        for row in csvFileReader:
            dates.append(int(row[0].split('-')[0]))
            prices.append(float(row[4]))
        
    return
```

```python
def predict_price(dates,prices,x):
    dates = np.reshape(dates,(len(dates),1))
    
    svr_lin = SVR(kernel='linear',C=1e3)
    svr_poly = SVR(kernel='poly', C=1e3,degree=2)
    svr_rbf = SVR(kernel='rbf',C=1e3,gamma='auto')
    
    svr_lin.fit(dates,prices)
    svr_poly.fit(dates,prices)
    svr_rbf.fit(dates,prices)
    
   
    plt.scatter(dates,prices,color='darkorange',label='Original Data')
    plt.plot(dates,svr_rbf.predict(dates),color='navy',label='RBF Model')
    plt.plot(dates,svr_poly.predict(dates),color='c',label='Polynomial Model')
    plt.plot(dates,svr_lin.predict(dates),color='cornflowerblue',label='Linear Model')
    
    plt.xlabel('Dates')
    plt.ylabel('Prices')
    plt.title('Support Vector Regression')
    plt.legend()
    plt.show()
    
    print('RBF Estimation :',svr_rbf.predict(x)[0])
    print('Linear Estimation :',svr_lin.predict(x)[0])
    print('Polynomial Estimation :',svr_poly.predict(x)[0])
    return 
```

```python
get_data('infy.csv')
predict_price(dates,prices,31)
```
